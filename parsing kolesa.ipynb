```python
import pandas as pd
import time
from bs4 import BeautifulSoup
import math
import urllib.request
```

```python
url = 'https://kolesa.kz/cars/'

cars = []
count = 2
while count <= 3:
    new_url = url+ '/?page=' + str(count)
    print(count)
    response = urllib.request.urlopen(new_url)
    soup = BeautifulSoup(response, 'html.parser')
    elements = soup.find_all('div', class_="a-list__item")
    vip_cards = soup.find_all('div', class_='vip-block__item vip-card js__a-card')
    cars.extend(elements)
    cars.extend(vip_cards)
    time.sleep(1)
    count += 1
```

```python
data = []
d = {}
for car in cars:
    try:
        title = car.find('h5', class_='a-card__title').text.strip()
        link = 'https://kolesa.kz' + car.a.get('href')
        print(title, link)
        respond = urllib.request.urlopen(link)
        html = BeautifulSoup(respond, 'html.parser')
        time.sleep(3)
        price = html.find('div', class_="offer__price").text.strip()
        
        info = {'Название' : title,
                'Цена' : price,
                'Ссылка': link}
        
        for data in html.find_all("dl"):
            key = data.dt.text.strip()
            value = data.dd.text.strip()
            print('{}: {}'.format(key,value))
        d[key] = value
        info.update(d)
        data.append(info)
    except:
        pass
```
